```python
import pandas as pd

df = pd.read_csv('datasets/testbed/TrainingDATA_Sep24.csv')


import holoviews as hv
hv.extension('bokeh')

temp_diff_avg = []
for i in range(8):
    temp_diff_avg.append(df['R{}_Temp'.format(i+1)] - df['R{}_Setpiont'.format(i+1)])
temp_diff_avg = sum(temp_diff_avg) / 8


hv.Curve(
    pd.DataFrame({
        'power': df['AHU_Power'],
        'temp_diff': temp_diff_avg,
    }).sort_values('temp_diff'),
    kdims=['temp_diff'], vdims=['power']
)
```
